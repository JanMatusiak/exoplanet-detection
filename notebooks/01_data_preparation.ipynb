### Data preparation

````python
koi = load_koi_full()
k2p = load_k2p_full()
print(f"KOI dataset shape: {koi.shape}")
print(f"K2P dataset shape: {k2p.shape}")
````

koi_full dataset and a k2p_full_default datasets will be used for analysis. koi_full dataset will be grouped by ```kepid``` and used for training, validating and testing, while k2p_full_default has already been filtered for duplicates and will only be used for testing. toi_full will not be used, as it contains less features than the other two datasets, making it difficult to cross-analyse them.  
19 physical features have been chosen for analysis based on their meritorical significance. Their exact meaning and description can be found in the README of the project. These features will be further inspected for missingness and correlation.

````python
koi_columns_set = list(KOI_PHYSICAL_COLUMNS_SET)
k2p_columns_set = list(K2P_PHYSICAL_COLUMNS_SET)
````

````python
koi1 = koi[koi_columns_set]
k2p1 = k2p_default[k2p_columns_set]
````

Columns ```kepi``` and ```pl_name``` will be used to group the data, while columns ```koi_disposition``` and ```disposition``` will contitute our labels. For KOI, FALSE POSITIVE objects will be mapped as 0, while CONFIRMED objects will be mapped as 1. For K2P, FALSE POSITIVE + REFUTED objects will be mapped as 0, while CONFIRMED will be mapped as 1. The rest will be extracted as production data, to be tested upon model deployment. Features with more than 50% of the values missing in the training set will be dropped. For the rest, a median imputer will be applied.

Mapping column names to uniform ones:

````python
koi2 = koi1.rename(columns=KOI_RENAME_MAP)
k2p2 = k2p1.rename(columns=K2P_RENAME_MAP)
````

Extracting desired labels:

````python
koi_labeled, koi_candidates = split_labeled_and_candidates(koi2)
k2p_labeled, k2p_candidates = split_labeled_and_candidates(k2p2)
````

Mapping labels:

````python
koi_labeled = map_labels(koi_labeled)
k2p_labeled = map_labels(k2p_labeled)
````

Converting units (required for Transit Depth - for KOI it is in ppm, for K2P it is in %):

````python
k2p_labeled = convert_transit_depth_percent_to_ppm(k2p_labeled)
````

Drop missing values upfront:

````python
koi_labeled.info()
````

````python
k2p_labeled.info()
````

````python
koi_labeled.drop(list(BASE_DROP_COLUMNS), axis=1, inplace=True)
k2p_labeled.drop(list(BASE_DROP_COLUMNS), axis=1, inplace=True)
````

### Split to train and test set

````python
X = koi_labeled.drop(columns=["group_id", "label"]).copy()
y = koi_labeled["label"]
groups = koi_labeled["group_id"]
````

````python
koi_train_set, koi_test_set = stratified_group_train_test_split(
    koi_labeled,
    n_splits=5,
    random_state=42,
    fold_index=0,
)
````
